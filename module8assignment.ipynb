QUES-2->Explain the difference between Absolute and Relative positioning?<br><br>
ANS=><br><br>
POSITION :RELATIVE=<br>.it is claculated relative to the element;s normal position in the document flow. <br>.the element still occupies it's original space ,even if it's visually moved. <br>.we can move it using top,right,bottom,left.<br><br>
ABSOLUTE POSITIONING:.it is calculated relative to the nearest positioned ancestor<br>.if no such ancestor is removed from the normal document flow,so it does not affect or reverse space for other elements<br>.if no ancestor exists,it is positioned relative to the html.

QUES-5->Explain the z-index, with a code example?<BR><BR>
ANS=><br>
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <style>
        .box{
            width: 200px;
            height: 200px;
            position: absolute;
        }
        .red{
            background-color: red;
            top:40px;
            left:40px;
            z-index:1;
        }
        .green{
            background-color: green;
            top:80px;
            left:80px;
            z-index:2;
        }
        .blue{
            background-color:blue;
            top:120px;
            left:120px;
            z-index:3;
        }

    </style>
</head>
<body>
    <div class="box red">z-index:1</div>
    <div class="box green">z-index:2</div>
    <div class="box blue">z-index:3</div>
</body>
</html>
<!--The  z-index property contrils the stacking order of elements along the z-axis(front to back direction).Elements with a higher z-index appear in front of those with alower one.-->
